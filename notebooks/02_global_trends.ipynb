```python
import sys
from pathlib import Path

project_root = Path().resolve().parent
sys.path.insert(0, str(project_root))

import matplotlib.pyplot as plt
from src.data_loader import load_crop_folder
```

```python
rice = load_crop_folder("../data/rice")
maize = load_crop_folder("../data/maize")
wheat = load_crop_folder("../data/wheat")
soybean = load_crop_folder("../data/soybean")
```

Global mean

```python
rice_g = rice.mean(dim=("lat", "lon"))
maize_g = maize.mean(dim=("lat", "lon"))
wheat_g = wheat.mean(dim=("lat", "lon"))
soybean_g = soybean.mean(dim=("lat", "lon"))
```

Time series

```python
plt.figure(figsize=(10,5))
rice_g.plot(label="Rice")
maize_g.plot(label="Maize")
wheat_g.plot(label="Wheat")
soybean_g.plot(label="Soybean")
plt.legend()
plt.title("Global mean crop yields (19812016)")
plt.grid(True)
plt.show()
```
